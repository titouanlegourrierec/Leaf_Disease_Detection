```python
for key,value in data.items():
    filename = value["filename"]
    
    img_path = f"{img_dir}/{filename}"
    
    print(img_path)
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    h,w,_ = img.shape

    mask = np.full((h,w), 255, dtype=np.uint8)
    
    regions = value["regions"]
    for region in regions:
        print(region)

        shape_attributes = region["shape_attributes"]
        region_attributes = region["region_attributes"]

        name = region_attributes["name"]

        if shape_attributes["name"] == "polygon":
            x_points = shape_attributes["all_points_x"]
            y_points = shape_attributes["all_points_y"]

            contours = []
            for x,y in zip(x_points,y_points):
                contours.append((x,y))
            
            contours=np.array(contours)
        elif shape_attributes["name"] == "circle":
            center_x = int(shape_attributes["cx"])
            center_y = int(shape_attributes["cy"])
            radius = int(shape_attributes["r"])

            contours = cv2.ellipse2Poly((center_x,center_y), (radius,radius), 0, 0, 360, 1)
        
        print(name)
        if name == "bg":
            color = 63
        elif name == "yellow_rust":
            color = 255
        elif name == "oidium":
            color = 191
        elif name == "leaf":
            color = 127

        cv2.drawContours(mask, [contours], -1, color, -1)

    filename_without_extension, extension = os.path.splitext(filename)
    new_filename = f"{filename_without_extension}_ground_truth{extension}"

    cv2.imwrite(f"{mask_dir}/{new_filename}", mask)
```
