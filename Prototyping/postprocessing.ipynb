```python
import os
import cv2
import numpy as np

def process_images(input_dir, output_dir):
    # Get a list of all image files in the directory
    image_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')]

    # Define the kernels for erosion and dilation
    kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
    kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (6,6))

    # Apply the function to each image file
    for image_file in image_files:
        # Read the image
        img = cv2.imread(os.path.join(input_dir, image_file), cv2.IMREAD_COLOR)

        # Apply dilation followed by erosion
        dilation = cv2.dilate(img, kernel1, iterations = 7) # 7
        erosion = cv2.erode(dilation, kernel2, iterations = 6) # 6

        # Save the modified image with the same filename in the output directory
        cv2.imwrite(os.path.join(output_dir, image_file), erosion)
```
